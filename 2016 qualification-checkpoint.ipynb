```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup complete')
```

```python
lines = open("busy_day.in","r").readlines()
R,K,D,T,L = [int(item) for item in lines[0].split()]
P = int(lines[1].strip())
products = np.array(list(map(int, lines[2].split())))
W = int(lines[3].strip())
ware_prod = []
warehouses = []
temp = 2*W + 4
for i in range(4,temp):
    if i%2 == 0:
        warehouses.append([int(item) for item in lines[i].split()])
    else:
        ware_prod.append(list(map(int, lines[i].split())))
C = int(lines[temp].strip())
order_coord = []
num_products_of_order = []
products_of_order = []
for i in range(temp + 1,len(lines)):
    if (i - temp) % 3 == 1:
        order_coord.append([int(item) for item in lines[i].split()])
    elif (i - temp) % 3 == 2:
        num_products_of_order.append(int(lines[i]))
    else:
        products_of_order.append(list(map(int, lines[i].split())))
```

```python
table = np.zeros([R,K])
#print(table.shape)
for i in range(C):
    x = order_coord[i][0]
    y = order_coord[i][1]
    table[x][y] += 1
for item in warehouses:
    x = item[0]
    y = item[1]
    table[x][y] = 4
```
